# Have games become better? An analysis of game reviews from 2010 to 2021

## Importing libraries and inspecting data

```python
import pandas as pd
import seaborn as sns
import numpy as np
import sklearn
from matplotlib import pyplot as plt
#import sqlite3
```

```python
reviews = pd.read_csv("all.csv")
```

```python
reviews.info()
```

```python
reviews.head(10)
```

As we can see, the dataset consistst of 11692 rows and 11 columns. Some rows are duplicate representing multiple values for developer, genre, platform and publisher for a given game. We also have some missing values for score and number of reviews. 

Since our focus is to analyse review scores our first order of priority is to drop rows that do not contain a score. 

```python
clean_reviews = reviews.dropna(subset=["score"]).drop(columns=["id"])
```

```python
clean_reviews.info()
```

After dropping rows without reviews we are left with 8339 rows and 11 columns. However, we can see that our releasedate column is not a datetime object column. Since we want to examine reviews over time we need to convert release_date to a datetime column. 

```python
clean_reviews["release_date"] = pd.to_datetime(clean_reviews["release_date"], format="%b %d %Y", errors="coerce")
```

```python
clean_reviews
```

## Examing reviews over time

```python
years = clean_reviews["release_date"].dt.year
years = years.dropna()
unique_years = years.unique().astype(int)

# Define figure and axis and use Seaborn to plot a lineplot representing review scores over years
fig, ax = plt.subplots(figsize=(10, 3))
lplot = sns.lineplot(ax=ax, data=clean_reviews, x=clean_reviews["release_date"].dt.year, y="score")
lplot.set(xticks=unique_years, title="Average review scores over time", xlabel="Release Year", ylabel="Average review scores")
```

```python
sns.set(rc={'figure.figsize':(30,8.27)})
sns.set_style("darkgrid")
sns.despine()
sns.scatterplot(data=clean_reviews, x="release_date", y="score", hue="genre", size="genre")

```
